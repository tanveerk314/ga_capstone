```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
metrics = pd.read_csv('data/eval_metrics.csv')
```

```python
metrics
```

```python
metrics.columns
```

```python
from matplotlib.pyplot import figure

figure(figsize=(16, 8),dpi=80 )
plt.plot(metrics['Epoch '],metrics['HasAns_f1'],label='HasAns_f1')
plt.plot(metrics['Epoch '], metrics['NoAns_f1'],label='NoAns_f1')
plt.plot(metrics['Epoch '],metrics['f1'],label='overall f1')
plt.title('Eval F1')
plt.xlabel('Epochs')
plt.ylabel('F1')
plt.legend()
```

```python
from matplotlib.pyplot import figure

figure(figsize=(16, 8), dpi=80)
plt.plot(metrics['Epoch '],metrics['NoAns_exact'],label='NoAns_exact')
plt.plot(metrics['Epoch '], metrics['HasAns_exact'],label='HasAns_exact')
plt.plot(metrics['Epoch '],metrics['exact'],label='exact match')
plt.title('Eval Exact Match')
plt.xlabel('Epochs')
plt.ylabel('Exact Match')
plt.legend()
```
